```python
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pandas
from sklearn.cluster import KMeans
import numpy as numpy


iris = datasets.load_iris()

# Transform dataset into pandas
# so we can have a look
df = pandas.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
# Clustering the data, x takes
# The first two columns of the
# dataset (called predictors)
X = iris.data[:, :]

# Assign the target to the label column
y = iris.target

# Visualise a scatter plot using X0 and X1
plt.scatter(X[:,0], X[:,1], c=y, cmap='gist_rainbow')
plt.xlabel('Sepa1 Length', fontsize=18)
plt.ylabel('Sepal Width', fontsize=18)
```

```python
# Applying the clustering algorithm K-means
# Clusters: Amount of center points placed
# n_jobs: Jobs to use for computation
# Random_state: Random number for generation
km = KMeans(n_clusters = 3, n_jobs = 4, random_state=21)
km.fit(X)


# Determine centerpoints of data 
# and return as array. 
# Each ROW is a cluster
# Each column is x, y..
centers = km.cluster_centers_

print(km.labels_)

```

```python
# Visualise the actual and predicted scatterplot

new_labels = km.labels_


# Plot the identified clusters and compare with the answers
fig, axes = plt.subplots(1, 2, figsize=(16,8))
axes[0].scatter(X[:, 0], X[:, 1])
axes[1].scatter(X[:, 0], X[:, 1], c=new_labels, cmap='jet')

axes[0].set_xlabel('Sepal length', fontsize=18)
axes[0].set_ylabel('Sepal width', fontsize=18)
axes[1].set_xlabel('Sepal length', fontsize=18)
axes[1].set_ylabel('Sepal width', fontsize=18)
axes[0].tick_params(direction='in', length=10, width=5, colors='k', labelsize=20)
axes[1].tick_params(direction='in', length=10, width=5, colors='k', labelsize=20)
axes[0].set_title('Actual', fontsize=18)
axes[1].set_title('Predicted', fontsize=18)
```
